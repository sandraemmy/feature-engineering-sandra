**Check for missing values**

```python
spaceship.isnull()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship['Cabin'] = spaceship['Cabin'].astype(str).str[0]
valid_decks = {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}
spaceship['Cabin'] =spaceship['Cabin'].apply(lambda x: x if x in valid_decks else None)
spaceship
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(columns=['PassengerId', 'Name'], errors='ignore').dropna()
spaceship
```

```python
spaceship.isna().sum()
```

- For non-numerical columns, do dummies.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False)
non_numerical_cols = spaceship.select_dtypes(exclude=['number']).columns
ohe.fit(spaceship[non_numerical_cols])
spaceship_trans_np = ohe.transform(spaceship[non_numerical_cols])
spaceship_trans_np
```

**Perform Train Test Split**

```python
if 'Transported' in spaceship.columns:
    x = spaceship.drop(columns=['Transported'])
    y = spaceship['Transported'].dropna()
    
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```
